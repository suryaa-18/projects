```python
from ultralytics import YOLO

# Load trained model
model = YOLO("runs/detect/train/weights/best.pt")

metrics = model.val()

# Extract Precision and Recall
precision = metrics.box.p.mean()  # Average Precision across all classes
recall = metrics.box.r.mean()  # Average Recall across all classes

# Compute Accuracy using F1-score formula
accuracy = (2 * precision * recall) / (precision + recall)

# Convert to percentage
precision_percent = precision * 100
recall_percent = recall * 100
accuracy_percent = accuracy * 100

print(f"Precision: {precision_percent:.2f}%")
print(f"Recall: {recall_percent:.2f}%")
print(f"Estimated Accuracy (F1 Score): {accuracy_percent:.2f}%")
```

```python
# Extract precision, recall, mAP@0.5, and mAP@0.5:0.95 as mean values
precision = metrics.box.p.mean() * 100  # Precision (mean over classes) in percentage
recall = metrics.box.r.mean() * 100     # Recall (mean over classes) in percentage
mAP_50 = metrics.box.map50.mean() * 100  # mAP@0.5 in percentage
mAP_50_95 = metrics.box.map.mean() * 100  # mAP@0.5:0.95 in percentage

# Print precision, recall, and mAP scores
print(f"Precision: {precision:.2f}%")
print(f"Recall: {recall:.2f}%")
print(f"mAP@0.5: {mAP_50:.2f}%")
print(f"mAP@0.5:0.95: {mAP_50_95:.2f}%")

# Calculate accuracy (if true positives, false positives, and false negatives are available)
try:
    true_positives = metrics.box.tp.sum()  # True positives
    false_positives = metrics.box.fp.sum()  # False positives
    false_negatives = metrics.box.fn.sum()  # False negatives

    total_predictions = true_positives + false_positives + false_negatives
    accuracy = (true_positives / total_predictions) * 100 if total_predictions > 0 else 0  # Avoid division by zero

    print(f"Accuracy: {accuracy:.2f}%")
except AttributeError:
    print("True positives, false positives, or false negatives are not available.")
```
